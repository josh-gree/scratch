```python
import pandas as pd
```

```python
columns = ['region_fuel','val']
data = [
    ('A prov', 100),
    ('coal',50),
    ('oil',50),
    ('B prov', 100),
    ('coal',20),
    ('oil',80)
]

shit_df1 = pd.DataFrame(data=data,columns=columns)


data = [
    ('A prov', 10),
    ('coal',5),
    ('oil',5),
    ('B prov', 10),
    ('coal',2),
    ('oil',8)
]

shit_df2 = pd.DataFrame(data=data,columns=columns)
```

```python
shit_df1['region'] = shit_df1.region_fuel.where(shit_df1.region_fuel.str.contains('prov'),None).ffill()
shit_df1
```

```python
shit_df1 = shit_df1[~shit_df1.region_fuel.str.contains('prov')]
shit_df1 = shit_df1.rename({'region_fuel':'fuel'},axis=1)
shit_df1
```

```python
shit_df2['region'] = shit_df2.region_fuel.where(shit_df2.region_fuel.str.contains('prov'),None).ffill()
shit_df2
```

```python
shit_df2 = shit_df2[~shit_df2.region_fuel.str.contains('prov')]
shit_df2 = shit_df2.rename({'region_fuel':'fuel'},axis=1)
shit_df2
```

```python
pd.merge(shit_df1,shit_df2,left_on=['fuel','region'],right_on=['fuel','region'])
```
